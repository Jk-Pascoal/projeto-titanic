```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Y = np.cos(X)
```

Criando a Figura e os Eixos

```python
fig, ax = plt.subplots()
```

Plotando os Dados

```python
# Cria a figura e os eixos
fig, ax = plt.subplots()

# Plota os dados
ax.plot(X, Y)

# Adiciona os rtulos e o ttulo
ax.set_title("Onda de Cosseno")
ax.set_xlabel("Valores do Eixo X")
ax.set_ylabel("Valores do Eixo Y")

# Exibe o grfico
plt.show()
```
